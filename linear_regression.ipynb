```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X_numpy, y_numpy = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=1)
# X_numpy, y_numpy

X_train = np.random.sample(size=(100, 1))
w_train = np.random.sample(size=(1, 1))
b_train = np.random.sample(size=1)
noise = 20
y_train = np.dot(X_train, w_train) + b_train * np.random.sample(size=(100, 1)) * 2
plt.scatter(X_train, y_train)
b_train
```

```python
n_samples, n_features = X_train.shape
input_size = n_features
output_size = 1
model = nn.Linear(input_size, output_size)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
# model.parameters()
```

```python
# X_train = torch.from_numpy(X_train)
# X_train = X_train.to(torch.float32)
# y_train = torch.from_numpy(y_train)
# y_train = y_train.to(torch.float32)
for i in range(100):
    y_predicted = model(X_train)
    loss = criterion(y_predicted, y_train)
    if not i % 10:
        print(loss)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
plt.scatter(X_train, y_train)
plt.plot(X_train, y_predicted.detach())
```
